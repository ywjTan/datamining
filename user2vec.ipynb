```python
import sys
import numpy as np
import pandas as pd
```

```python
users = pd.read_csv('./data/userinfo.csv')
```

```python
movievec=pd.read_csv('./data/movie_vec.csv')
```

```python
user_list=[]
for i in range(users.shape[0]):
    user_dict={}
    user_dict['movies'] = users['movies'][i].strip('[').strip(']').split(',')
    user_dict['userId'] = users['userId'][i]
    user_dict['vec'] = np.array([0 for i in range(128)])
    user_list.append(user_dict)
```

```python
movie_dict={}
for i in range(movievec.shape[0]):
    movie_dict[movievec['movieId'][i]]=list(map(float,movievec['movieVec'][i].strip('[').strip(']').split(',')))
```

```python
for i in range(len(user_list)):
    count=0
    for item in user_list[i]['movies']:
        if item=='':
            continue
        if movie_dict.__contains__(int(item)):
            count+=1
            user_list[i]['vec']=user_list[i]['vec']+np.array(movie_dict[int(item)])
    print(count)
    if count!=0:
        user_list[i]['vec']=user_list[i]['vec']/count
```

```python
userId=[]
userVec=[]
for item in user_list:
    userId.append(item['userId'])
    userVec.append(item['vec'].tolist())
```

```python
dataframe = pd.DataFrame({'userId': userId, 'userVec': userVec})
```

```python
dataframe.to_csv("./data/user_vec.csv", index=False, sep=',')
```
